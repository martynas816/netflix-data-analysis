```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

# Preview data
df.head()
```

```python
df['type'].value_counts()

# Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', title='Movies vs TV Shows')
```

There are far more movies than TV shows on Netflix.
This suggests Netflix historically Netflix has published a higher quantity of movies, even though TV shows drive long-term engagement.

```python
# Top 5 genres/categories
df['listed_in'].value_counts().head(5).plot(kind='barh', title='Top Genres')
```

The most frequent categories include:

Dramas

Documentaries

Stand-up Comedy


This suggests Netflix invests heavily in:

Drama to reach global audiences

Documentaries for educational niche content

Stand-Up for entertainment & lower production cost

```python
# Content by year
df['release_year'].value_counts().sort_index().plot(kind='line', title='Content by Year')
```
